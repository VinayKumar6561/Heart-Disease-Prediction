```python
# IMPORTING LIBRARIES
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# DATA COLLECTION AND PROCESSING
```

```python
# LOADING DATASET

heart_data = pd.read_csv('heart.csv')
```

```python
heart_data.head(7)
```

```python
heart_data.info()
```

```python
heart_data.describe()
```

```python
 # 0 --> Healthy Heart
 # 1 --> UnHealthy Heart
```

```python
# SPLITTING THE FEATURES AND THE TARGET
```

```python
X=heart_data.iloc[:,:-1]
```

```python
X
```

```python
Y=heart_data[['target']]
```

```python
Y
```

```python
# SPLITTING DATA INTO TRAINING AND TESTING DATA
```

```python
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.30,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# MODEL TRAINING

# LOGISTIC REGRESSION
```

```python
model=LogisticRegression()
```

```python
# TRAIN MODEL WITH TRAINING DATA
```

```python
model.fit(X_train,Y_train)
```

```python
# MODEL EVALUATION (ACCURACY SCORE)
```

```python
X_train_predict= model.predict(X_train)

train_data_accuracy=accuracy_score(X_train_predict,Y_train)
```

```python
print('Accuracy on Training Data', train_data_accuracy)
```

```python
X_test_predict= model.predict(X_test)

test_data_accuracy=accuracy_score(X_test_predict,Y_test)
```

```python
print('Accuracy on Test Data', train_data_accuracy)
```

```python
# BUILDING PREDICTIVE SYSTEM
```

```python
input_data=(49,1,1,130,266,0,1,171,0,0.6,2,0,2)

input_data=np.asarray(input_data).reshape(1,-1)

prediction=model.predict(input_data)
print('Prediction : ',prediction)

if prediction[0]==0:
    print(' The person does not have a Heart Disease')
else:
     print(' The person have a Heart Disease')
```
